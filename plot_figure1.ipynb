```python
# Figure 1-2 share a raster plot from an example session
import scipy.io as sio
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
plt.style.use('scifigs.mplstyle')
SAVE_FIG = False
```

```python
# load example session
subject = 'satchel'
mat_fname = 'preprocessed_data/all_data_delay_{}.mat'.format(subject)
ex_sess = 'Sa191224'
dat = sio.loadmat(mat_fname,squeeze_me=True,struct_as_record=False)['all_data']
curr_dat = getattr(dat, ex_sess)

# get example trial counts
left_idx, right_idx = curr_dat.arr.LH_idx.astype(bool), curr_dat.arr.RH_idx.astype(bool)
counts = curr_dat.ex_spike_times # n_neurons x n_bins
left_counts, right_counts = np.flipud(counts[left_idx,:]), np.flipud(counts[right_idx,:])
```

```python
fig,ax = plt.subplots(1,2, constrained_layout=True, sharey=True)
fig.set_figwidth(2*fig.get_figwidth())

ax[0].scatter(np.where(left_counts)[1],np.where(left_counts)[0],marker='|',color='k')
ax[1].scatter(np.where(right_counts)[1],np.where(right_counts)[0],marker='|',color='k')

if SAVE_FIG:
    pdf = PdfPages('figs/example_raster.pdf')
    pdf.savefig(fig)
    pdf.close()
else:
    fig.show()
```
